```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger le fichier classeur
# Charger le fichier Excel
df = pd.read_excel("PROD.xlsx")
```

```python
# Convert 'Date' column to datetime format
df['BelegDat'] = pd.to_datetime(df['BelegDat'])

# Group by 'Désignation'
groups = df.groupby('Material')

# Define a function to calculate the difference in months between consecutive dates
def calculate_date_diff(group):
    group = group.sort_values('BelegDat')
    group['DateDiff'] = group['BelegDat'].diff().dt.days / 30
    return group

# Apply the function to each group and update the 'Statu' column based on the condition
for name, group in groups:
    group = calculate_date_diff(group)
    nan_statu_indices = group[group['Artikle Zustand'].isna()].index
    group.loc[nan_statu_indices, 'Status'] = 'Slow Mover'
    group.loc[group['DateDiff'] < 6, 'Status'] = 'Fast Mover'
    df.update(group)
```

```python
import pandas as pd

# Handle missing values in "Artikle Zustand" (optional)
# ... (replace this line with your logic to address missing values)

# Ensure BelegDat has a datetime format (if necessary)
if not pd.api.types.is_datetime64_dtype(df['BelegDat']):
  try:
    # Attempt conversion with potential format errors set to 'coerce'
    df['BelegDat'] = pd.to_datetime(df['BelegDat'], errors='coerce')
  except:
    # Handle potential conversion errors (e.g., invalid format)
    print("Error converting BelegDat to datetime format.")

# Group by 'Material'
groups = df.groupby('Material')

# Define a function to process and calculate differences
def calculate_differences(group):
  # Sort by BelegDat
  group = group.sort_values('BelegDat')

  # Forward fill missing dates (optional)
  # This assumes it's okay to use the previous date for missing values
  # Consider alternative handling based on your data
  group['BelegDat'] = group['BelegDat'].fillna(method='ffill')

  # Calculate date differences (handle potential NaNs)
  group['DateDiff'] = group['BelegDat'].diff().dt.days / 30
  group['DateDiff'] = group['DateDiff'].fillna(0)  # Replace NaNs with 0 (optional)

  return group

# Apply the function to each group and update the 'Status' column
for name, group in groups:
  group = calculate_differences(group)

  # Assign statuses based on conditions (handle potential NaNs)
  group.loc[group['Artikle Zustand'].isna(), 'Artikle Zustand'] = 'Slow Mover'
  group.loc[group['DateDiff'] < 6, 'Artikle Zustand'] = 'Fast Mover'

  # Update the main DataFrame
  df.update(group)
```
