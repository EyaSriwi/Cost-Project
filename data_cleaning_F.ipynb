# Load the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Data
df = pd.read_excel("Production.xlsx")
df.head()
```

```python
# Convert 'BelegDat' to datetime format with correct format
df['BelegDat'] = pd.to_datetime(df['BelegDat'], format='%d.%m.%Y', errors='coerce')
```

# Handling Missing Values

```python
# Check for missing values in each column
missing_values = df.isna().sum()

# Calculate the percentage of missing values in each column
missing_percentage = (missing_values / len(df)) * 100

# Combine the missing values count and percentage into a DataFrame
missing_data_info = pd.DataFrame({'Missing Values': missing_values, 'Missing Percentage': missing_percentage})

print(missing_data_info)
```

```python
# Rename columns '      Menge' and '      Nettowert'
df = df.rename(columns={'      Menge': 'Menge', '     Nettowert': 'Nettowert','   Gesamtvbrwert':'Gesamtvbrwert'})
df
```

# Data Visualization

```python
# Select numerical columns
num_columns = df.select_dtypes(include=['number']).columns

# Calculate the number of plots and the number of rows needed
num_plots = len(num_columns)
num_rows = (num_plots // 3) + (num_plots % 3 > 0)

# Define figure size and individual subplot size
fig_width = 10
fig_height = 5
subplot_width = 3
subplot_height = 2

# Create subplots with the calculated layout
fig, axes = plt.subplots(num_rows, 3, figsize=(fig_width, fig_height))

# Flatten the axes array to simplify iteration
axes = axes.flatten()

# Plot each numerical column
for i, column in enumerate(num_columns):
    ax = axes[i]
    df[column].plot.hist(ax=ax)
    ax.set_title(column)
    ax.grid(True)

# Hide empty subplots
for i in range(num_plots, num_rows * 3):
    fig.delaxes(axes[i])

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
plt.show()
```

```python

# Pair plot for numerical columns
sns.pairplot(df.select_dtypes(include=['number']))
plt.show()
```

```python
# Summary statistics for "Nettowert"
print(f'Standard Deviation of Nettowert: {df["Nettowert"].std()}')
print(f'Mean of Nettowert: {df["Nettowert"].mean()}')
```

```python
# Distribution plot for "Nettowert"
plt.figure(figsize=(8, 6))
sns.distplot(df['Nettowert'], kde=True)
plt.title('Distribution plot of Nettowert')
plt.show()
```

```python
# Correlation matrix for numerical columns only
numeric_df = df.select_dtypes(include=['number'])
corr = numeric_df.corr()
print(corr)
```

```python
# Heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
